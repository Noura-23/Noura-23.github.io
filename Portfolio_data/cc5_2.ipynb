```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Scrape the skyscraper center website
url = "https://www.skyscrapercenter.com/buildings"

# Get the webpage
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Try using pandas to read tables
tables = pd.read_html(url)
df = tables[0]

#Check actual column names
print(df.columns.tolist())

# The column names are really long, so let's find the ones we need
# We'll select columns that START WITH these words
name_col = [col for col in df.columns if col.startswith('Name')][0]
city_col = [col for col in df.columns if col.startswith('City')][0]
height_col = [col for col in df.columns if col.startswith('Height')][0]

# Select just those columns
df_clean = df[[name_col, city_col, height_col]].copy()

# Rename to simple names
df_clean.columns = ['building_name', 'city', 'height']

# Remove rows with missing data
df_tidy = df_clean.dropna()

# Keep top 20
df_tidy = df_tidy.head(21)

# Display to verify
print(df_tidy)
```

```python
import pandas as pd

# Read the existing CSV
df = pd.read_csv('tallest_buildings.csv')

# Extract just the meter value by splitting at the space
df['height_m'] = df['height'].str.replace('\xa0', ' ').str.split(' ').str[0].astype(int)

# Keep only the columns we want
df_final = df[['building_name', 'city', 'height_m']]

# Save the new CSV
df_final.to_csv('tallest_buildings.csv', index=False)

# Display result
print(df_final)
```
