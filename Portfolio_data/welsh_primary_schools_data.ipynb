```python
import pandas as pd
```

```python
df = pd.read_csv('state_primary_schools_wales.csv')
```

```python
schools_clean = df[['name', 'lat', 'lon', 'language', 'town_name']].copy()
df
```

```python
schools_clean = schools_clean.dropna(subset=['lat', 'lon'])
```

```python
schools_clean = schools_clean.dropna(subset=['name'])
```

```python
schools_clean = schools_clean.rename(columns={
    'name': 'School_Name',
    'lat': 'latX',
    'lon': 'longY',
    'language': 'Language',
    'town_name': 'Town'
})
```

```python
schools_clean.to_csv('welsh_primary_schools_clean.csv', index=False)
```
