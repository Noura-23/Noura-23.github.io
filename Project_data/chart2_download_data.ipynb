```python
import pandas as pd
import requests
```

```python
from pandas.core.frame import DataFrame
# API configuration
base_url = "https://api.worldbank.org/v2/country"
indicator = "EG.USE.ELEC.KH.PC" #Electric power consumption (kWh per capita) - world bank
year = "2022"

african_countries = [
    'DZA', 'AGO', 'BEN', 'BWA', 'CMR', 'TCD', 'CIV', 'EGY', 'ETH',
    'GAB', 'GHA', 'KEN', 'LBY', 'MAR', 'MOZ', 'NAM', 'NER', 'NGA', 'RWA',
    'SEN', 'ZAF', 'TGO','TUN', 'ZWE', 'ZMB'
]

#Join countries for API
countries_string = ";".join(african_countries)

#Build API url
api_url = f"{base_url}/{countries_string}/indicator/{indicator}?date={year}&format=json&per_page=500"

#display
print(api_url)

#fetch data
response = requests.get(api_url)
if response.status_code == 200:
    data = response.json()

    if len(data) > 1 and data[1]:
        records = []

        for item in data[1]:
            if item['value'] is not None:
                records.append({
                    'country': item['country']['value'],
                    'iso': item['countryiso3code'],
                    'consumption': item['value'],
                    'year': item['date']
                })

        df = pd.DataFrame(records)

       # Sort by consumption (ascending for chart)
        df = df.sort_values('consumption', ascending=False)
        print(df)
```

```python
 #save data
 df.to_csv('chart2.csv', index=False)
 print("\n Saved: chart2_consumption.csv")
```
