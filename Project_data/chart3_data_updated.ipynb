```python
import pandas as pd
```

```python
# Load your current chart3_final_data.csv (with 2022 data)
current_data = pd.read_csv('https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Project_data/chart3_final_data.csv')

print("Current data (2022):")
print(current_data.head())
print(f"Shape: {current_data.shape}")
```

```python
# Load your NEW 2024 population data
# Replace this URL with your actual 2024 population CSV link
population_2024 = pd.read_csv('https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Project_data/population.csv', skiprows=4)  # Adjust skiprows if needed
population_2024 = population_2024.rename(columns={'Country Code': 'iso', '2024': 'Population_2024'})
population_2024 = population_2024[['iso', 'Population_2024']]

print("\n2024 Population data:")
print(population_2024.head())
```

```python
# Load your NEW 2024 GDP per capita PPP data
# Replace this URL with your actual 2024 GDP CSV link
gdp_2024 = pd.read_csv('https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Project_data/gdp_per_capita_ppp.csv', skiprows=4)  # Adjust skiprows if needed
gdp_2024 = gdp_2024.rename(columns={'Country Code': 'iso', '2024': 'GDP_2024'})
gdp_2024 = gdp_2024[['iso', 'GDP_2024']]

print("\n2024 GDP data:")
print(gdp_2024.head())
```

```python
# Merge 2024 population data
updated_data = current_data.merge(population_2024, on='iso', how='left')

# Merge 2024 GDP data
updated_data = updated_data.merge(gdp_2024, on='iso', how='left')

# Replace old columns with new data
updated_data['Population'] = updated_data['Population_2024']
updated_data['GDP per capita, PPP'] = updated_data['GDP_2024']

# Drop the temporary columns
updated_data = updated_data.drop(columns=['Population_2024', 'GDP_2024'])

print("\nUpdated data (2024):")
print(updated_data.head())
print(f"Shape: {updated_data.shape}")

# Check for any missing values
print("\nMissing values check:")
print(updated_data.isnull().sum())

# Save the updated dataset
updated_data.to_csv('chart3_final_data_2024.csv', index=False)
print("\n✓ Saved as 'chart3_final_data_2024.csv'")

# # Download in Colab
# from google.colab import files
# files.download('chart3_final_data_2024.csv')
```
