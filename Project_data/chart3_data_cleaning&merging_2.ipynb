```python
import pandas as pd
```

```python
region_df=pd.read_csv("https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/refs/heads/master/all/all.csv")
region_df=region_df[["alpha-3", "region", "sub-region"]]
region_df
region_df = region_df.rename(columns={"alpha-3": "iso"})
region_df
```

```python
tariffs_df=pd.read_csv("https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Project_data/chart3_tariffs_gp_scraped.csv")
tariffs_df
tariffs_df=tariffs_df[["iso", "tariff_usd_per_kwh"]]
tariffs_df=tariffs_df.rename(columns={"country": "name", "tariff_usd_per_kwh": "tariffs"})
tariffs_df
```

```python
raw_links = ["https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Project_data/population.csv", "https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Project_data/gdp_per_capita_ppp.csv"]
titles = ["Population", "GDP per capita, PPP"]
toskip = [4, 0]

cleaned_dfs = []

for i in range(2):
  df = pd.read_csv(raw_links[i], skiprows=toskip[i])
  df = df.rename(columns={"Country Code": "iso", "Country Name": "name", "2022": titles[i]})
  df = df[["name", "iso", titles[i]]]
  df = df.merge(region_df, on="iso")
  df = df.merge(tariffs_df, on="iso")
  df = df.query("region == 'Africa'")
  cleaned_dfs.append(df)

# merge the cleaned final dfs into a wide format
df = pd.merge(cleaned_dfs[0], cleaned_dfs[1], on=["name", "region", "iso", "sub-region"])

# Clean up: remove duplicate tariff columns
df = df.drop(columns=["tariffs_y"])
df = df.rename(columns={"tariffs_x": "tariffs"})

df
```

```python
# Save the cleaned dataset
df.to_csv("chart3_final_data.csv", index=False)
```
