```python
import pandas as pd
```

```python
region_df=pd.read_csv("https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/refs/heads/master/all/all.csv")
region_df=region_df[["alpha-3", "region"]]
region_df
region_df = region_df.rename(columns={"alpha-3": "iso"})
region_df
```

```python
raw_links = ["https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Week%209/Access%20to%20electricity.csv", "https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Week%209/Electric%20power%20consumption.csv", "https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Week%209/Population.csv", "https://raw.githubusercontent.com/Noura-23/Noura-23.github.io/refs/heads/main/Week%209/GDP%20per%20capita.csv"]
titles = ["Electricity Access", "Electric Power Consumption", "Population", "GDP per Capita"]
toskip = [0,4, 4,4]

cleaned_dfs = []

for i in range(4):
  df=pd.read_csv(raw_links[i], skiprows=toskip[i])
  df = df.rename(columns={"2022 [YR2022]": titles[i], "2022": titles[i], "Country Name": "name", "Country Code": "iso"})
  df = df[["name", "iso", titles[i]]]
  df = df.dropna()
  df = df.merge(region_df, on="iso")
  df = df.query("region == 'Africa'")
  cleaned_dfs.append(df)

# merge the cleaned final dfs into a wide format
df = pd.merge(cleaned_dfs[0], cleaned_dfs[1], on=["name", "region", "iso"])
df = pd.merge(df, cleaned_dfs[2], on=["name", "region", "iso"])
df = pd.merge(df, cleaned_dfs[3], on=["name", "region", "iso"])
df
```
